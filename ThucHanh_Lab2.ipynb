```python
# ==============================================================================
# LAB 2: BÀI THỰC HÀNH TRÌNH BÀY VÀ TRỰC QUAN HÓA DỮ LIỆU
# Mã nguồn sử dụng file 'processed_dulieuxettuyendaihoc.csv' từ Lab 1
# ==============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Thiết lập cấu hình đồ họa cho trực quan hóa đẹp mắt
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 7)
plt.rcParams['font.size'] = 11

# Đọc file dữ liệu đã xử lý từ Lab 1
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# ==============================================================================
# PHẦN 1: THỐNG KÊ DỮ LIỆU
# ==============================================================================
print("====== PHẦN 1: THỐNG KÊ DỮ LIỆU ======\n")

# 1. Sắp xếp dữ liệu điểm DH1 tăng dần
df_sorted_dh1 = df.sort_values(by='DH1')

# 2. Sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
df_sorted_dh2_gt = df.sort_values(by=['GT', 'DH2'])

# Định nghĩa các hàm phân vị bổ sung cho Pivot Table
def q1(x): return x.quantile(0.25)
def q2(x): return x.quantile(0.50)
def q3(x): return x.quantile(0.75)

agg_funcs = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q2, q3]

# 3. Pivot-table của DH1 theo KT (Khối thi)
print("3. Pivot-table DH1 theo Khối Thi (KT):")
pivot_kt = df.pivot_table(values='DH1', index='KT', aggfunc=agg_funcs)
display(pivot_kt)

# 4. Pivot-table của DH1 theo KT và KV (Khu vực)
print("\n4. Pivot-table DH1 theo Khối Thi (KT) và Khu Vực (KV):")
pivot_kt_kv = df.pivot_table(values='DH1', index=['KT', 'KV'], aggfunc=agg_funcs)
display(pivot_kt_kv)

# 5. Pivot-table của DH1 theo KT, KV và DT (Dân tộc)
print("\n5. Pivot-table DH1 theo KT, KV và DT (Một vài dòng đầu):")
pivot_complex = df.pivot_table(values='DH1', index=['KT', 'KV', 'DT'], aggfunc=agg_funcs)
display(pivot_complex.head(10))


# ==============================================================================
# PHẦN 2: TRÌNH BÀY DỮ LIỆU
# ==============================================================================
print("\n====== PHẦN 2: TRÌNH BÀY DỮ LIỆU ======\n")

# 1. Trình bày dữ liệu biến GT
print("Tần số và Tần suất biến Giới tính (GT):")
freq_gt = df['GT'].value_counts()
pct_gt = df['GT'].value_counts(normalize=True) * 100
display(pd.DataFrame({'Tần số': freq_gt, 'Tần suất (%)': pct_gt}))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
df['GT'].value_counts().plot(kind='bar', ax=axes[0], color=['skyblue', 'pink'])
axes[0].set_title("Biểu đồ tần số Giới tính (GT)")
axes[0].set_ylabel("Số lượng học sinh")

df['GT'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[1], colors=['skyblue', 'pink'], startangle=90)
axes[1].set_title("Biểu đồ tần suất Giới tính (GT)")
plt.show()

# 2. Trình bày dữ liệu các biến US_TBM1, US_TBM2, US_TBM3
plt.figure(figsize=(10, 6))
df[['US_TBM1', 'US_TBM2', 'US_TBM3']].boxplot()
plt.title("Biểu đồ Box-plot so sánh phân bổ điểm thang Mỹ qua các năm")
plt.ylabel("Điểm số (Thang 4)")
plt.show()

# 3. Trình bày dữ liệu biến DT với học sinh là Nam (Giả định 'M' là Nam)
print("Thống kê dân tộc cho học sinh Nam:")
df_nam = df[df['GT'] == 'M']
display(df_nam['DT'].value_counts().to_frame(name='Số lượng'))

# 4. Trình bày dữ liệu biến KV với học sinh Nam, dân tộc Kinh (Ký hiệu là 1), điểm thỏa điều kiện
# Lưu ý: Điều chỉnh điều kiện lọc 'DT' == 1 nếu tệp gốc của bạn gán nhãn chuỗi khác
df_filter_4 = df[(df['GT'] == 'M') & (df['DT'] == 1) & (df['DH1'] >= 5.0) & (df['DH2'] >= 4.0) & (df['DH3'] >= 4.0)]
print("\nThống kê Khu vực (KV) thỏa điều kiện câu 4:")
display(df_filter_4['KV'].value_counts().to_frame(name='Số lượng'))

# 5. Trình bày dữ liệu các biến DH1, DH2, DH3 >= 5.0 và thuộc khu vực 2NT
df_filter_5 = df[(df['DH1'] >= 5.0) & (df['DH2'] >= 5.0) & (df['DH3'] >= 5.0) & (df['KV'] == '2NT')]
print(f"\nSố lượng học sinh khu vực 2NT đạt điểm tất cả các môn đại học từ 5.0 trở lên: {len(df_filter_5)} học sinh.")


# ==============================================================================
# PHẦN 3: TRỰC QUAN HÓA DỮ LIỆU THEO NHÓM PHÂN LOẠI
# ==============================================================================
print("\n====== PHẦN 3: TRỰC QUAN HÓA THEO NHÓM ======\n")

# 1. Học sinh nữ trên các nhóm XL1, XL2, XL3 dạng unstacked
df_nu = df[df['GT'] == 'F']
xl_matrix = df_nu[['XL1', 'XL2', 'XL3']].apply(pd.Series.value_counts).reindex(['Y', 'TB', 'K', 'G', 'XS']).fillna(0)
xl_matrix.T.plot(kind='bar', width=0.8)
plt.title("Xếp loại học lực của các học sinh Nữ qua 3 năm học")
plt.ylabel("Số lượng học sinh")
plt.xlabel("Các năm học (Lớp 10, 11, 12)")
plt.xticks(rotation=0)
plt.legend(title="Xếp loại")
plt.show()

# 2-7. Gom cụm trực quan hóa bằng đồ thị cột kép (Seaborn Countplot)
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Câu 3: Số lượng thí sinh từng khu vực dựa trên nhóm khối thi
sns.countplot(data=df, x='KT', hue='KV', ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title("Số lượng thí sinh từng Khu vực theo Khối thi")

# Câu 4: Số lượng thí sinh đậu/rớt trên từng nhóm khối thi
sns.countplot(data=df, x='KT', hue='KQXT', ax=axes[0, 1], palette='coolwarm')
axes[0, 1].set_title("Kết quả Đậu(1)/Rớt(0) dựa trên Khối thi")

# Câu 5: Số lượng thí sinh đậu rớt trên từng nhóm khu vực
sns.countplot(data=df, x='KV', hue='KQXT', ax=axes[1, 0], palette='coolwarm')
axes[1, 0].set_title("Kết quả Đậu(1)/Rớt(0) dựa trên Khu vực")

# Câu 7: Số lượng thí sinh đậu rớt trên từng nhóm giới tính
sns.countplot(data=df, x='GT', hue='KQXT', ax=axes[1, 1], palette='PRGn')
axes[1, 1].set_title("Kết quả Đậu(1)/Rớt(0) dựa trên Giới tính")

plt.tight_layout()
plt.show()


# ==============================================================================
# PHẦN 4: TRỰC QUAN HÓA DỮ LIỆU NÂNG CAO
# ==============================================================================
print("\n====== PHẦN 4: TRỰC QUAN HÓA NÂNG CAO ======\n")

# 1. Vẽ biểu đồ đường Simple cho biến T1
plt.figure(figsize=(10, 5))
df['T1'].value_counts().sort_index().plot(kind='line', marker='o', color='purple')
plt.title("Biểu đồ đường Simple thể hiện phân bổ tần số của điểm Toán lớp 10 (T1)")
plt.xlabel("Điểm số T1")
plt.ylabel("Tần số xuất hiện")
plt.show()

# 2. Tạo biến phân loại phanlopt1 cho môn Toán (T1)
def phan_lop_t1(score):
    if score < 5.0: return 'k'      # Kém
    elif score < 7.0: return 'tb'   # Trung bình
    elif score < 8.0: return 'kh'   # Khá (Dùng 'kh' để phân biệt kí tự)
    else: return 'g'                # Giỏi

df['phanlopt1'] = df['T1'].apply(phan_lop_t1)

# 3. Lập bảng tần số cho biến phanlopt1
print("Bảng tần số cho biến phanlopt1:")
display(df['phanlopt1'].value_counts().to_frame(name='Tần số'))

# 4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi phanlopt1
plt.figure(figsize=(12, 6))
t1_pivot = df.pivot_table(index='T1', columns='phanlopt1', aggfunc='size', fill_value=0)
t1_pivot.plot(kind='line', marker='s', ax=plt.gca())
plt.title("Biểu đồ Multiple Line cho điểm T1 phân theo nhóm Học lực")
plt.ylabel("Số lượng học sinh")
plt.xlabel("Điểm số T1")
plt.show()


# ==============================================================================
# PHẦN 5: MÔ TẢ DỮ LIỆU VÀ KHẢO SÁT DẠNG PHÂN PHỐI
# ==============================================================================
print("\n====== PHẦN 5: MÔ TẢ DỮ LIỆU VÀ DẠNG PHÂN PHỐI ======\n")

# 1. Mô tả thống kê, kiểm tra Skewness và Kurtosis của biến T1
print("Các chỉ số thống kê mô tả của biến T1:")
print(df['T1'].describe())
print(f"Độ lệch (Skewness): {df['T1'].skew():.4f}")
print(f"Độ nhọn (Kurtosis): {df['T1'].kurt():.4f}")

# Vẽ đồ thị phân phối (Histogram) và đồ thị hộp (Box-plot) cho T1
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.boxplot(x=df['T1'], ax=axes[0], color='lightgreen')
axes[0].set_title("Biểu đồ Box-plot của biến T1")

sns.histplot(df['T1'], kde=True, ax=axes[1], color='blue')
axes[1].set_title("Biểu đồ Histogram kèm đường mật độ (KDE) của T1")
plt.show()

# Kiểm chứng phân phối chuẩn bằng QQ-Plot
plt.figure(figsize=(6, 6))
stats.probplot(df['T1'], dist="norm", plot=plt)
plt.title("Biểu đồ QQ-Plot kiểm định phân phối chuẩn của biến T1")
plt.show()

# 3. Khảo sát tương quan giữa biến phụ thuộc DH1 theo biến độc lập T1
cov_val = df['DH1'].cov(df['T1'])
corr_val = df['DH1'].corr(df['T1'])
print(f"Hiệp phương sai (Covariance) giữa DH1 và T1: {cov_val:.4f}")
print(f"Hệ số tương quan tuyến tính Pearson (Correlation): {corr_val:.4f}")

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='T1', y='DH1', alpha=0.8, color='red')
sns.regplot(data=df, x='T1', y='DH1', scatter=False, color='black', label='Đường xu hướng')
plt.title("Biểu đồ tương quan Scatter-plot giữa điểm thi Đại học môn 1 (DH1) và T1")
plt.legend()
plt.show()

# 5. Khảo sát tương quan chéo giữa cụm biến DH1, DH2, DH3
corr_matrix = df[['DH1', 'DH2', 'DH3']].corr()
print("\nMa trận tương quan giữa các môn thi Đại học (DH1, DH2, DH3):")
display(corr_matrix)

plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".3f", vmin=-1, vmax=1)
plt.title("Ma trận nhiệt (Heatmap) thể hiện hệ số tương quan giữa DH1, DH2, DH3")
plt.show()

print("\n>>> HOÀN THÀNH TOÀN BỘ BÀI LAB 2! Biểu đồ đã được xuất đầy đủ thành công.")
```
